```python
from solid import *
from solid.objects import *
from solid.utils import *

def lego(l,w):
    h=7.5
    d1=4.8
    d2=7.5
    s=8
    
    l=int(l)
    w=int(w)
    
    if l<=1:
        i=l
        l=w
        w=i
        
    unit=cylinder(d=d2,h=h)
    brick=cube([(l-1)*s,(w-1)*s,h])
    if w>1:
        unit=scale([1,1,.5])(minkowski()(brick,unit))
    else:
        unit=hull()(unit,right(s*(l-1))(unit))
    j=0
    while j<w:
        k=0    
        while k<l:
            hole=back(-j*s)(down(1)(cylinder(d=d1,h=h+2)))
            hole=right(k*s)(hole)
            unit=unit-hole
            k+=1
        j+=1
    
    scad_render_to_file(unit,file_header=f'$fn = {100};')
    
```

```python
def nema(std):
    if std==8:
        a=16
        b=4
        d=15
        g=3
    if std==11:
        a=23
        b=5
        d=22
        g=4
    if std==14:
        a=26
        b=5
        d=22
        g=4
    if std==17:
        a=31
        b=5
        d=22
        g=.12*25.4
    if std==23:
        a=47.14
        b=6.35
        d=38.1
        g=5
    if std==34:
        a=69.7
        b=9.5
        d=73
        g=5.5
    if std==42:
        a=88.9
        b=16
        d=60
        g=6
    
    thick=b/2
    base=cube([std*2.54/2,std*2.54/2,thick])-translate([a/2,a/2,-1])(cylinder(d=g,h=thick+2))-cylinder(d=d+2,h=10,center=True)
    
    h=5.5
    d1=4.8
    d2=7.5
    s=8
    
    l=int(std/2.32)
    
    
    unit=cylinder(d=d2,h=h)
    unit=hull()(unit,right(s*(l-1))(unit))
    
    k=0    
    while k<l:
        hole=(down(1)(cylinder(d=d1,h=h+2)))
        hole=right(k*s)(hole)
        unit=unit-hole
        k+=1
    edge=translate([-s*(l-1)/2,std*2.54/2+d1/2,0])(unit)
    edge2=rotate([0,0,-90])(edge)
    corner=base+edge+edge2
    half=mirror([1,0,0])(corner)+corner
    final=mirror([0,1,0])(half)+half
    
    scad_render_to_file(final,file_header=f'$fn = {100};')
```

```python
def shaft(std):
    if std==8:
        a=16
        b=4
        d=15
        g=3
    if std==11:
        a=23
        b=5
        d=22
        g=4
    if std==14:
        a=26
        b=5
        d=22
        g=4
    if std==17:
        a=31
        b=5
        d=22
        g=.12*25.4
    if std==23:
        a=47.14
        b=6.35
        d=38.1
        g=5
    if std==34:
        a=69.7
        b=9.5
        d=73
        g=5.5
    if std==42:
        a=88.9
        b=16
        d=60
        g=6
    
    wall=2
    shaft=cylinder(d=b+wall,h=b*3)-down(1)(cylinder(d=b,h=b*3+2))
    
    
    h=5.5
    d1=4.8
    d2=7.5
    s=8
    
    l=3
    
    unit=cylinder(d=d2,h=h)
    unit=hull()(unit,right(s*(l-1))(unit))
    
    k=0    
    while k<l:
        hole=(down(1)(cylinder(d=d1,h=h+2)))
        hole=right(k*s)(hole)
        unit=unit-hole
        k+=1
    final=shaft+translate([-8,0,b*3])(unit)
    scad_render_to_file(final,file_header=f'$fn = {100};')
```
